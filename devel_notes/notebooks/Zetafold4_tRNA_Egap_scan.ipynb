```python
from alphafold.parameters import KT_IN_KCAL
from math import exp, log
#1ehz
sequence = 'GCGGAUUUAGCUCAGUUGGGAGAGCGCCAGACUGAAGAUCUGGAGGUCCUGUGUUCGAUCCACAGAAUUCGCACCA'
p = partition( sequence, params = params, mfe = True)
dG = -KT_IN_KCAL * log( p.Z )
```

```python
secstruct = '(((((((..((((........)))).(((((.......))))).....(((((.......))))))))))))....'
p = partition( sequence, params = params, mfe = True, structure = secstruct )
dG = -KT_IN_KCAL * log( p.Z )
```

```python
C_eff_stacked_pairs = [ 10**(0.2*x) for x in range(15*5) ]
dG_tot     = [None]*len(C_eff_stacked_pairs) 
struct_MFE = [None]*len(C_eff_stacked_pairs) 
dG_struct  = [None]*len(C_eff_stacked_pairs) 
for n,C_eff_stacked_pair in enumerate(C_eff_stacked_pairs):
    params = get_params()
    params.C_eff_stacked_pair = C_eff_stacked_pair
    p = partition( sequence, params = params, mfe = True)
    dG_tot[n] = -KT_IN_KCAL * log( p.Z )
    struct_MFE[n] = p.struct_MFE
    p = partition( sequence, params = params, mfe = True, structure = secstruct )
    dG_struct[n] = -KT_IN_KCAL * log( p.Z )
```

```python
dG_gap = []
for x,y in zip(dG_tot,dG_struct): dG_gap.append(x-y)
for mfe,val in zip(struct_MFE,C_eff_stacked_pairs): print mfe,val
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = np.array( [C_eff_stacked_pairs,dG_gap])
data = pd.DataFrame( data )
f, ax = plt.subplots(dpi=300)
g = sns.lineplot( x=C_eff_stacked_pairs,y=dG_gap,dashes=False,linewidth=1.5,ax=ax )
g.set_xscale( 'log' )
g.set(xlabel='C_eff_stacked_pair',ylabel='dG (kcal/mol)' )
```
